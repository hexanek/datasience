```python
import numpy as np
```

```python
from sklearn import datasets, linear_model
```

```python
from sklearn.model_selection import train_test_split
```

```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
boston = datasets.load_boston()
```

```python
f, axarr = plt.subplots(5,2, sharex = True, sharey= True, figsize=(12,12))
for i in range(0,5):
    for j in range(0,2):
        boston_X = boston.data[:,np.newaxis,i*2+j]
        boston_X_train = boston_X[:-20]
        
        boston_X_test = boston_X[-20:]
        
        boston_Y_train = boston.target[:-20]
        boston_Y_test = boston.target[-20:]
        
        regr = linear_model.LinearRegression()
        regr.fit(boston_X_train, boston_Y_train)
                
        axarr[i,j].scatter(boston_X_test, boston_Y_test,  color='red')
        axarr[i,j].plot(boston_X_test, regr.predict(boston_X_test), color='blue', linewidth=1)
        
plt.show()

        
    
```

```python
sluby = pd.read_csv('sluby.csv', sep=' ')
```

```python
sluby_X_train = sluby.ludnosc[:10]
sluby_X_test = sluby.ludnosc[10:]
sluby_Y_train = sluby.malzenstwa[:10]
sluby_Y_test = sluby.malzenstwa[10:]

regr = linear_model.LinearRegression()


regr.fit(sluby_X_train.values.reshape(-1,1), sluby_Y_train.values.reshape(-1,1))

plt.scatter(sluby_X_test.values, sluby_Y_test.values,  color='red')
plt.plot(sluby_X_test, regr.predict(sluby_X_test.values.reshape(-1,1)), color='blue', linewidth=1)
plt.title('Wykres regresji liniowej ilosci małżeństw do ludności')
plt.xlabel('Ludność')
plt.ylabel('Małżeństwa')
plt.show()
```

```python
wydatki_csv = pd.read_csv('wydatki.csv', sep=' ')
```

```python
dochody_X_train = wydatki_csv.dochody[:6]
dochody_X_test = wydatki_csv.dochody[6:]
wydatki_Y_train = wydatki_csv.wydatki[:6]
wydatki_Y_test = wydatki_csv.wydatki[6:]
```

```python
regr = linear_model.LinearRegression()


regr.fit(dochody_X_train.values.reshape(-1,1), wydatki_Y_train.values.reshape(-1,1))

plt.scatter(dochody_X_test.values, wydatki_Y_test.values,  color='red')
plt.plot(dochody_X_test, regr.predict(dochody_X_test.values.reshape(-1,1)), color='blue', linewidth=1)
plt.title('Wykres regresji liniowej ilości wydatków do dochodów')
plt.xlabel('Dochody')
plt.ylabel('Wydatki')
plt.show()
```

```python
numbers = np.random.uniform(0,1,size=(2,100)) 
avg_X = numbers[0,:].mean()
avg_Y = numbers[1,:].mean()
```

```python
avg_Y
```

```python
diff_X = np.array([x - avg_X for x in numbers[0,:]])
diff_Y = np.array([y - avg_Y for y in numbers[1,:]])
```

```python
diff_X
```

```python
diff_Y
```

```python
diff_Xdo2 = np.square(diff_X)
diff_Ydo2 = np.square(diff_Y)
```

```python
diff_Xdo2
```

```python
diff_Ydo2
```

```python
iloczyn = diff_Xdo2 * diff_Ydo2
```

```python
dane = pd.DataFrame(data = {'X': numbers[0,:],
                           'Y': numbers[1,:],
                           'X-X¯': diff_X,
                           'Y-Y¯': diff_Y,
                           'Kwadrat_roznic_X': diff_Xdo2,
                           'Kwadrat_roznic_Y': diff_Ydo2,
                           'Iloczyn_roznic': iloczyn})
```

```python
dane
```

```python
licznik = dane['Iloczyn_roznic'].sum()
```

```python
mianownik = dane['Kwadrat_roznic_X'].sum()
```

```python
a = licznik/mianownik
```

```python
b = avg_Y - a * avg_X
```
