```python
 import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
import pandas as pd
```

```python
fury = pd.read_csv('samochody1tys.csv')   #Pobranie danych
```

```python
x = np.linspace(0, 2, 100)

plt.plot(x, x, label='liniowa')
plt.plot(x, x**2, label='kwadratowa')
plt.plot(x, x**3, label='trzeciego rzędu')
plt.plot(x, np.sin(x), label='funkcji sinus')  #wykres z dodatkową funkcją sinus

plt.xlabel('x label')
plt.ylabel('y label')

plt.title("Wykres funkcji kwadratowych")

plt.legend()

plt.show()
```

```python
fury["markaa"] = fury["marka"].astype("category")
```

```python
dane=fury.groupby("markaa").size()  #pobranie rozmiaru każdej marki
```

```python
xx=fury["marka"]
```

```python
xlist=set(xx)  #pobranie wszystkich marek bez powtórzeń
```

```python
dnax=list(xlist)   #konwersja z setu do listy
```

```python
plt.bar(dnax,dane,label='liczebność')
#wykres słupkowy marka i ich liczebność
plt.xlabel('marka')
plt.ylabel('liczebność')

plt.title("Wykres liczebności aut")

plt.legend()

plt.show()
```

```python
fury["silnik"] = fury["rodzaj_silnika"].astype("category")
```

```python
ysilnik=fury.groupby("silnik").size() #pobranie liczebności danych rodzajów silników
```

```python
ysilnik
```

```python
xsilnik=['Benzyna','Benzyna+CNG','Benzyna+LPG','Diesel','Elektryczny','Hybryda'] #utworznie osi x
```

```python
plt.pie(ysilnik,labels=xsilnik)
#wykres kołowy rodzaj silnika i ich liczebność

plt.title("Wykres użytkowania danego rodzaju silnika")

plt.legend()

plt.show()
```
